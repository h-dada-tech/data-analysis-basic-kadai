```python
# matplotlib.pyplotメソッドとjapanize_matplotlibメソッドとnumpyをインポート
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib

# nrows=2, ncols=3で、6領域のサブプロットを作成
fig, axes = plt.subplots(2, 3, tight_layout=True)

# axes[0,0]に折れ線グラフを表示する
y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)

# データを作成
# 0から10までの幅で配列を等間隔に100個作成する
x = np.linspace(0, 10, 100)
y = 2 + 2 * np.sin(2 * x)

axes[0,1].plot(x, y, linewidth=2.0)
axes[0,1].set(xlim=(0, 10), xticks=np.arange(0, 10),
       ylim=(-1, 5), yticks=np.arange(-1, 6))

# 平均が15, 標準偏差5の標準正規分布から2000個の乱数の生成する
x = np.random .normal(15, 5, 2000)
# ヒストグラムをプロット
axes[0,2].hist(x)

# データを生成する
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 60)
y = 4 + np.random.normal(0, 2, len(x))

# プロットする
axes[1,0].scatter(x, y)

# 日本語タイトルを付与する。
axes[0,0].set_title("折れ線グラフ")
axes[0,1].set_title("sin関数")
axes[0,2].set_title("ヒストグラム")
axes[1,0].set_title("散布図")

plt.show()
```
